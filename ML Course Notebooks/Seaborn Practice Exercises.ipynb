```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
sns.jointplot(x='fare',y='age',data=df)
```

```python
sns.distplot(df['fare'],kde=False, bins=30,color='red')
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(x='class',y='age',data=df,palette='rainbow')
```

```python
sns.swarmplot(x='class',y='age',data=df,palette='plasma')
```

```python
sns.countplot(df['sex'],palette='coolwarm')
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(),cmap='coolwarm')
```

```python
g = sns.FacetGrid(data=df,col='sex')
g = g.map(plt.hist,'age')
```
