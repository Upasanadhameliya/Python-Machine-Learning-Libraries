```python
df['reason']=df['title'].apply(lambda x:x.split(":")[0])
```

```python
df['reason'].value_counts()
```

```python
sns.countplot(df['reason'],palette='Set2')
```

```python
type(df['timeStamp'][0])
```

```python
df['timeStamp']=pd.to_datetime(df['timeStamp'])
```

```python
df.head()
```

```python
df.info()
```

```python
df['hour']=df['timeStamp'].apply(lambda x:x.hour)
```

```python
df['month']=df['timeStamp'].apply(lambda x:x.month)
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['day of week']=df['timeStamp'].apply(lambda x:dmap[x.weekday()])
```

```python
df.head()
```

```python
sns.set_context('talk')
plt.figure(figsize=(10,5))
sns.countplot(df['day of week'],hue=df['reason'],palette='plasma')
plt.legend(loc=(1.04,0.8))
```

```python
sns.set_context('talk')
plt.figure(figsize=(10,5))
sns.countplot(df['month'],hue=df['reason'],palette='plasma')
plt.legend(loc=(1.04,0.8))
```

```python
gb = df.groupby('month')
```

```python
countdf = gb.count()
```

```python
countdf['month'] = countdf.index
```

```python
countdf.head()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df['month'].value_counts())
```

```python
sns.set_style('whitegrid')
sns.set_context('notebook')
sns.lmplot(data=countdf,x='month',y='lat')
```
