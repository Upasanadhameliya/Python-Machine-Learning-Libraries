```python
print('Supplies Subgroup: ',sales_data['Supplies Subgroup'].unique(),'\n')
print('Supplies Group: ',sales_data['Supplies Group'].unique(),'\n')
print('Region: ',sales_data['Region'].unique(),'\n')
print('Route To Market: ',sales_data['Route To Market'].unique(),'\n')
print('Opportunity Result: ',sales_data['Opportunity Result'].unique(),'\n')
print('Competitor Type: ',sales_data['Competitor Type'].unique(),'\n')
```

```python
sales_data['Supplies Subgroup'] = le.fit_transform(sales_data['Supplies Subgroup'])
sales_data['Supplies Group'] = le.fit_transform(sales_data['Supplies Group'])
sales_data['Region'] = le.fit_transform(sales_data['Region'])
sales_data['Route To Market'] = le.fit_transform(sales_data['Route To Market'])
sales_data['Opportunity Result'] = le.fit_transform(sales_data['Opportunity Result'])
sales_data['Competitor Type'] = le.fit_transform(sales_data['Competitor Type'])
```

```python
sales_data.head()
```

```python
sales_data.dtypes
```

```python
cols = [col for col in sales_data.columns if col not in ['Opportunity Number','Opportunity Result']]
data = sales_data[cols]
target = sales_data['Opportunity Result']
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
data_train,data_test,target_train,target_test = train_test_split(data,target,test_size=0.30,random_state = 10)
# here data_train and target_train are the training data and the training target(output)
# which are used for training, here both the data and the target would be fed to the system

# based on the training data and target, the system will be capable of predicting on its own
# for that we would input data_test and we would get output as target_test
# i.e. we would input test data and would be presented with a test target(output) that would be 
# predicted by the system
data_test['Supplies Subgroup'].count()
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
gnb = GaussianNB()
pred = gnb.fit(data_train,target_train).predict(data_test)

print("Naive-Bayes Accuracy: ",accuracy_score(target_test,pred,normalize=False))

# 17768 / 23408 = 0.7590567327409433
print("Naive-Bayes Accuracy: ",accuracy_score(target_test,pred,normalize=True))
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

```python
svc_model = LinearSVC(random_state=0)
pred = svc_model.fit(data_train,target_train).predict(data_test)
print("LinearSVC accuracy: ",accuracy_score(target_test,pred,normalize=True))
# here what actually happens is that target_test consists of actual outputs
# pred model consists of predicted outputs so accuracy_score compares the actual output
# and the predicted output and gives the score based on the number of matches that
# were true
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
neigh = KNeighborsClassifier(n_neighbors = 3)
neigh.fit(data_train,target_train)
pred = neigh.predict(data_test)
print("KNeighbors Accuracy: ",accuracy_score(target_test,pred,normalize=True))
```
