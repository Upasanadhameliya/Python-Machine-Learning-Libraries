```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
img2 = cv2.resize(img2,(600,600))
```

```python
plt.imshow(img2)
```

```python
x_offset = img1.shape[1] - img2.shape[1]
y_offset = img1.shape[0] - img2.shape[0]
```

```python
roi = img1[y_offset:y_offset+img2.shape[0],x_offset:x_offset+img2.shape[1]]
```

```python
plt.imshow(roi)
```

```python
mask = cv2.cvtColor(img2,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(mask,cmap="gray")
```

```python
mask = cv2.bitwise_not(mask)
```

```python
plt.imshow(mask,cmap="gray")
```

```python
mask.shape
```

```python
white_bk = np.full(mask.shape,255,dtype='uint8')
```

```python
mask = cv2.bitwise_or(white_bk,white_bk,mask=mask)
```

```python
plt.imshow(mask,cmap="gray")
```

```python
mask.shape
```

```python
img2 = cv2.bitwise_or(img2,img2,mask=mask)
```

```python
plt.imshow(img2)
```

```python
roi = cv2.bitwise_or(roi,img2)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(roi)
```

```python
img1[y_offset:y_offset+roi.shape[0],x_offset:x_offset+roi.shape[1]] = roi
```
