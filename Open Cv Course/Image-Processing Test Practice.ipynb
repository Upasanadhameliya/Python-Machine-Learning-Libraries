```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
def loadimg():
    img = cv2.imread("../DATA/giraffes.jpg")
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img
```

```python
img = loadimg()
```

```python
display_img(img)
```

```python
ret, thresh = cv2.threshold(cv2.cvtColor(img,cv2.COLOR_RGB2GRAY),127,255,cv2.THRESH_BINARY)
```

```python
display_img(thresh,"gray")
```

```python
img = loadimg()
```

```python
display_img(img)
```

```python
display_img(cv2.cvtColor(img,cv2.COLOR_RGB2HSV))
```

```python
kernel = np.ones((4,4),dtype='uint8')/10
```

```python
img = loadimg()
```

```python
display_img(img)
```

```python
blurred = cv2.filter2D(img,-1,kernel)
```

```python
display_img(blurred)
```

```python
sobelx = cv2.Sobel(cv2.cvtColor(img,cv2.COLOR_RGB2GRAY),cv2.CV_64F,1,0,ksize=5)
```

```python
display_img(sobelx,"gray")
```

```python
img = cv2.imread("../DATA/giraffes.jpg")
```

```python
colors = ('b','g','r')
for i,colr in enumerate(colors):
    hist = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist,color=colr)
    
plt.title("Giraffes Histograms")
```
