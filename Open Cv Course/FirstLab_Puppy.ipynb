```python
# The last three images are darker than the above three
# because in the last three all the three dimensions are present
# and the other two are made to be 0 completely
# and 0 stands for black color.. thats why two layers of black color
# make it appear darker

# while in the above three images, there is just only one dimension
# which is achieved by slicing.. and the other two dimensions are not shown
# and hence because the black isn't accumulating, the images are lighter

# to get the original image back we would get it by combining the first three images
# together, because if we combine the last three images it would have 2 + 2 + 2 i.e.
# 6 dimensions of black, i.e. it would be 50 shades darker.. hehe
```

```python
import cv2
import numpy as np
```

```python
drawing = False
ix,iy = -1,-1


def draw_rekt(event,x,y,flag,params):
    global drawing,ix,iy
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix,iy = x,y
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing:
            cv2.rectangle(img,pt1=(ix,iy),pt2=(x,y),color=(255,127,99),thickness=2)
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.rectangle(img,pt1=(ix,iy),pt2=(x,y),color=(255,127,99),thickness=2)
        
cv2.namedWindow('rekt')

cv2.setMouseCallback('rekt',draw_rekt)

img = np.zeros((512,512,3),dtype='int8')
while True:
    cv2.imshow('rekt',img)
    
    if cv2.waitKey(1) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```
