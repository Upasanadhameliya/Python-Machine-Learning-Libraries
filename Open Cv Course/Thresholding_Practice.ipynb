```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("DATA/rainbow.jpg",0)
```

```python
def showimg(img):
    plt.imshow(img,cmap="gray")
```

```python
showimg(img)
```

```python
ret, img1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
showimg(img1)
```

```python
ret, img1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
```

```python
showimg(img1)
```

```python
ret, img1 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
```

```python
showimg(img1)
```

```python
ret, img1 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
```

```python
showimg(img1)
```

```python
ret, img1 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)
```

```python
showimg(img1)
```

```python
news = cv2.imread("DATA/crossword.jpg",0)
```

```python
def newsshow(img):
    fig = plt.figure(figsize=(18,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap="gray")
```

```python
newsshow(news)
```

```python
ret, img1 = cv2.threshold(news,150,255,cv2.THRESH_BINARY)
```

```python
newsshow(img1)
```

```python
img2 = cv2.adaptiveThreshold(news,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,3,6)
```

```python
newsshow(img2)
```

```python
# LATER ON TRY SOME NOISE REMOVAL TECHNIQUES ON THE ABOVE PHOTO AND GET A BETTER NEWS READING
```

```python
img3 = cv2.adaptiveThreshold(news,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,3,6)
```

```python
newsshow(img3)
```

```python
combined = cv2.addWeighted(src1=img1,alpha=0.5,src2=img3,beta=0.5,gamma=0)
```

```python
newsshow(combined)
```
