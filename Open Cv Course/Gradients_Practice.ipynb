```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def loadimg():
    img = cv2.imread("DATA/sudoku.jpg",0)
    return img
```

```python
def showimg(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap="gray")
```

```python
img = loadimg()
showimg(img)
```

```python
sobelx = cv2.Sobel(img,5,1,0,5)
sobely = cv2.Sobel(img,5,0,1,5)
laplacian = cv2.Laplacian(img,5,ksize=5)
```

```python
showimg(sobelx)
```

```python
showimg(sobely)
```

```python
showimg(laplacian)
```

```python
lapblur = cv2.medianBlur(laplacian,5)
```

```python
showimg(lapblur)
```

```python
blended = cv2.addWeighted(src1=sobelx,alpha=0.5,src2=sobely,beta=0.5,gamma=0)
```

```python
showimg(blended)
```

```python
lapblur = cv2.medianBlur(laplacian,3)
```

```python
showimg(lapblur)
```

```python
ret, thresh = cv2.threshold(lapblur,160,255,cv2.THRESH_BINARY_INV)
```

```python
showimg(thresh)
```

```python
kernel = np.ones((2,2),dtype='uint8')
```

```python
out = cv2.morphologyEx(thresh,cv2.MORPH_CLOSE,kernel)
```

```python
showimg(out)
```

```python
gradientmorph = cv2.morphologyEx(blended,cv2.MORPH_GRADIENT,kernel)
```

```python
ret, thresh1 = cv2.threshold(gradientmorph,20,255,cv2.THRESH_BINARY_INV)
```

```python
showimg(thresh1)
```

```python
out1 = cv2.morphologyEx(thresh1,cv2.MORPH_CLOSE,kernel)
```

```python
showimg(out1)
```
