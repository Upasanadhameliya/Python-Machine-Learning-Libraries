```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../DATA/dog_backpack.jpg')
```

```python
type(img)
```

```python
plt.imshow(img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
plt.imshow(cv2.flip(img,0))
```

```python
imcpy = img.copy()
```

```python
cv2.rectangle(imcpy,pt1=(200,360),pt2=(600,700),color=(255,0,0),thickness=5)
```

```python
plt.imshow(imcpy)
```

```python
imcpy = img.copy()
```

```python
vertices = np.array([[200,700],[600,700],[350,360]])
```

```python
vertices = vertices.reshape(-1,1,2)
```

```python
cv2.polylines(imcpy,pts=[vertices],isClosed=True,color=(0,0,255),thickness=5)
```

```python
plt.imshow(imcpy)
```

```python
imcpy = img.copy()
```

```python
cv2.fillPoly(imcpy,pts=[vertices],color=(0,0,255))
```

```python
plt.imshow(imcpy)
```

```python
greybk = np.zeros((512,512,3),dtype='int8')

def empty_rekt(event,x,y,flag,params):
    if event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(greybk,center=(x,y),radius=50,color=(120,120,215))
        
cv2.namedWindow('Draw_Here')
cv2.setMouseCallback('Draw_Here',empty_rekt)

while True:
    cv2.imshow('Draw_Here',greybk)
    
    if cv2.waitKey(1) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```
