```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def loadimg():
    img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img,"ABCDE",(50,300),font,5,(255,255,255),23,cv2.LINE_AA)
    return img
```

```python
img = loadimg()
```

```python
def showimg(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap="gray")
```

```python
showimg(img)
```

```python
kernel = np.ones((5,5),dtype='uint8')
```

```python
img = cv2.erode(img,kernel)
```

```python
showimg(img)
```

```python
img = loadimg()
```

```python
showimg(img)
```

```python
img = cv2.erode(img,kernel,iterations= 5)
```

```python
showimg(img)
```

```python
img = loadimg()
```

```python
showimg(img)
```

```python
whitenoise = np.random.randint(low=0,high=2,size=(600,600))
```

```python
whitenoise *= 255
```

```python
img = img + whitenoise
```

```python
showimg(img)
```

```python
img = cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel)
```

```python
showimg(img)
```

```python
img = cv2.medianBlur(img.astype('float32')/255,5)
```

```python
showimg(img)
```
