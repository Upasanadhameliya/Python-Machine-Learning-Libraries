```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
licensedetector = cv2.CascadeClassifier("../DATA/haarcascades/haarcascade_licence_plate_rus_16stages.xml")
```

```python
carimg = cv2.imread("../DATA/car_plate.jpg")
```

```python
plt.figure(figsize=(12,10))
plt.imshow(carimg)
```

```python
def whatdoesitdo(img):
    imcpy = img.copy()
    coord = licensedetector.detectMultiScale(imcpy)
    
    for (x,y,w,h) in coord:
        cv2.rectangle(imcpy,(x,y),(x+w,y+h),(0,255,0),2)
    return imcpy
```

```python
result = whatdoesitdo(carimg)
```

```python
plt.figure(figsize=(12,10))
plt.imshow(result)
```

```python
def blurplate(img):
    imcpy = img.copy()
    coord = licensedetector.detectMultiScale(imcpy)
    
    for (x,y,w,h) in coord:
        plteim = imcpy[y:y+h,x:x+w]
        plteim = cv2.medianBlur(plteim,9)
        imcpy[y:y+h,x:x+w] = plteim
        
    return imcpy
```

```python
result = blurplate(carimg)
```

```python
plt.figure(figsize=(12,10))
plt.imshow(result)
```
