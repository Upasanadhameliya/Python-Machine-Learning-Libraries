```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("pokemon_data.csv",index_col=0)
```

```python
df.head()
```

```python
sns.lmplot(x='Attack',y='Defense',data=df)
```

```python
sns.lmplot(x='Attack',y='Defense',data=df,fit_reg=False,hue='Generation')
```

```python
sns.lmplot(x='Attack',y='Defense',data=df,fit_reg=False,hue='Generation')
plt.xlim(0,None)
plt.ylim(0,None)
```

```python
sns.boxplot(data=df)
```
